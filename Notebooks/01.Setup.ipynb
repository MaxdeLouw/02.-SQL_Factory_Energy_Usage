```python
import pandas as pd
import sqlite3 as sqlite
from pathlib import Path
```

```python
raw_path = "..\Data\Raw\Steel_industry_data.csv"
db_path = "../Data/Factory_Energy_Usage.db"

df = pd.read_csv(raw_path)
conn = sqlite.connect(db_path)

# Convert the date column to datetime format
df["date"] = pd.to_datetime(df["date"], dayfirst=True)

# Convert tons of CO2 to kg of CO2
df['CO2(kg)'] = df['CO2(tCO2)'] * 1000

df.to_sql("energy_usage", conn, if_exists="replace", index=False);
conn.close()
```

```python
conn = sqlite.connect(db_path)

tables = { 
    "KPIs",
    "dashboard_hourly",
    "dashboard_weekday",
    "dashboard_month",
    "dashboard_load_type",
    "energy_usage_clean"
}

for table in tables:
    df = pd.read_sql(f"SELECT * FROM {table}", conn)
    df.to_csv(f"../Data/Processed/{table}.csv", index=False)
    print(f"Exported {table} to CSV.")

conn.close()
```
